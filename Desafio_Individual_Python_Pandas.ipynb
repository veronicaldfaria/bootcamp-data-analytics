```python
# 1. Ao visualizar a base você percebeu que seria melhor alterar o nome de
# algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da coluna
# que indica a pressão sanguínea, mude a coluna ‘Ocupação’ para ‘Profissão’,
# a coluna ‘Categoria BMI’ está em parte em inglês, substitua para ‘Categoria
# IMC’.

renomear_colunas = {
    "ID": "Identificador",
    "Ocupação": "Profissão",
    "Categoria BMI": "Categoria IMC",
    "Pressão sanguíneaaaa": "Pressão sanguínea"
}
df.rename(columns=renomear_colunas, inplace=True)
df.head()
```

```python
# 2. Qual é a média, a moda e a mediana de horas de sono para cada uma
# das profissões? [‘mean’, np.median, pd.Series.mode]

media = df.groupby("Profissão")["Duração do sono"].mean()
mediana = df.groupby("Profissão")["Duração do sono"].median()
moda = df.groupby("Profissão")["Duração do sono"].apply(lambda x: x.mode().values[0] if not x.mode().empty else np.nan)
estatisticas_sono = pd.DataFrame({
    'Média': media,
    'Mediana': mediana,
    'Moda': moda
})
estatisticas_sono
```

```python
# 3. Das pessoas que atuam com engenharia de software qual a
# porcentagem de obesos?

eng_software = df[df["Profissão"] == "Eng. de Software"]
total_eng_software = len(eng_software)
obesos = np.sum(eng_software["Categoria IMC"] == "Obesidade")
porcentagem_obesos = (obesos / total_eng_software) * 100
porcentagem_obesos
```

```python
# 4. De acordo com os dados, advogar ou ser representante de vendas faz
# você dormir menos? (Use o método ‘isin’, considere a média)

profissoes = ["Advogado", "Representante de Vendas"]
df_filtrado = df[df["Profissão"].isin(profissoes)]
media_sono = np.mean(df_filtrado["Duração do sono"])
media_sono
```

```python
# 5. Entre quem fez enfermagem e quem fez medicina, quem tem menos
# horas de sono? (Use o método ‘isin’, considere a média)

cursos = ["Enfermagem", "Medicina"]
df_filtrado_cursos = df[df["Profissão"].isin(cursos)]
media_sono_med_enf = np.mean(df_filtrado_cursos["Duração do sono"])
media_sono_med_enf
```

```python
# 6. Faça um subconjunto com as colunas Identificador, Gênero, Idade,
# Pressão sanguínea e Frequência cardíaca.

subconjunto = df[["Identificador", "Gênero", "Idade", "Pressão sanguínea", "Frequência cardíaca"]]
subconjunto.head()
```

```python
# 7. Descubra qual a profissão menos frequente no conjunto. (Use
# value_counts)

profissao_menos_frequente = df["Profissão"].value_counts().idxmin()
profissao_menos_frequente
```

```python
# 8. Quem tem maior pressão sanguínea média, homens ou mulheres?
# (Considere a média)

df["Pressão sistólica"] = df["Pressão sanguínea"].str.split("/").str[0].astype(float)
df["Pressão diastólica"] = df["Pressão sanguínea"].str.split("/").str[1].astype(float)
media_pressao_por_genero = df.groupby("Gênero")[["Pressão sistólica", "Pressão diastólica"]].mean()
media_pressao_por_genero
```

```python
# 10. Pessoas com frequências cardíacas acima de 70 dão mais passos que
# pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

# 1. Filtrando as pessoas com frequência cardíaca acima de 70 e calculando a média de passos
passos_acima_70 = df[df["Frequência cardíaca"] > 70]["Passos diários"].mean()

# 2. Exibindo a média de passos para pessoas com FC acima de 70
print(f"Média de passos para FC > 70: {passos_acima_70:.2f}")
```
