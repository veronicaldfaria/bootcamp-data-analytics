# Desafio 4: InsectSprays

Definição do problema: Dado que você está trabalhando num
laboratório de inseticidas. Você precisa fazer uma pesquisa em todos
os sprays e verificar qual jogo é o mais eficaz.

--------------

Código do dataset: InsectSpray

--------------

Dicas:
- Use o método.idmax

```python
InsectSprays = pydataset.data("InsectSprays")
InsectSprays.head()
```

**1. Calcule o número total de insetos mortos (count) por tipo de spray
(spray).**


```python
insetos_Mortos = InsectSprays.groupby("spray")["count"].sum().reset_index()
insetos_Mortos
```

**2. Identifique qual spray foi o mais eficaz em matar insetos.**

```python
#insetos_Mortos.idxmax()
insetos_Mortos.loc[insetos_Mortos["count"].idxmax()]
```

**3. Calcule a média geral e filtre os dados para mostrar apenas sprays
com eficácia acima da média.**

```python
media = InsectSprays['count'].mean() #puxa a media geral
print(f"A media geral é: {media}") #sintaxe de print
```

```python
acimaMedia = InsectSprays[InsectSprays['count'] > media] #faz comparação
acimaMedia
```

# Desafio 5: Chicktws

Definição do problema: Foi conduzido um experimento para medir
e comparar a eficácia de vários suplementos alimentares na taxa
de crescimento de galinhas.
Os pintinhos recém-nascidos foram alocados aleatoriamente em
seis grupos, e cada grupo recebeu um suplemento alimentar
diferente. Seus pesos em gramas após seis semanas são fornecidos
junto com os tipos de ração.

------------

Código do dataset: chickwts

------------

Dicas:
- Método .reset_index()
- Método idxmax()
- Método .groupby()



```python
chickwts = pydataset.data("chickwts")
```

```python
chickwts.head()
```

**1. Calcule o peso médio dos pintinhos (weight) para cada tipo de
ração (feed).**

```python
peso_medio= chickwts.groupby("feed")["weight"].mean().reset_index()
peso_medio
```

**2. Identifique a ração que resultou no maior peso médio.**

```python
#maior_peso = peso_medio.loc[peso_medio["weight"].idxmax()]
maior_peso = peso_medio["weight"].idxmax()
maior_peso
```

**3. Crie uma tabela que mostre o peso máximo, mínimo e médio
para cada tipo de ração.**

```python
tabela = chickwts.groupby("feed")["weight"].agg([min,max,np.mean]).reset_index()
tabela
```

# Desafio 6: Attitude



Definição do problema: De uma pesquisa com funcionários
administrativos de uma grande organização financeira, os dados
são agregados dos questionários de aproximadamente 35
funcionários para cada um dos 30 departamentos (selecionados
aleatoriamente). Os números dão a proporção percentual de
respostas favoráveis a sete perguntas em cada departamento.

--------------

**Dicionário:**

**rating:** nota geral do departamento

**complaints:** fator categórico sobre reclamações

**privileges:** acesso a benefícios

**learning:** fator categórico sobre aprendizado

**raises:** fator categórico sobre aumentos salariais

**critical:** fator categórico sobre feedback crítico

**advance:** oportunidades de crescimento


--------------
Código do dataset: Attitude

--------------
Dica:
- Método .reset_index()
groupby

```python
attitude = pydataset.data("attitude")
```

```python
attitude.head()
```

**1 - Calcule a média de "aprendizado" (learning) por nível de motivação (rating).**


```python
media = attitude.groupby("rating")["learning"].mean().reset_index()
#media = attitude.groupby("rating")["learning"].mean()
media
```

**2 - Encontre a combinação de rating e complaints que possui o maior valor médio de aprendizado.**

```python
combinação = attitude.loc[attitude["learning"].idxmax(),["rating","complaints"]]
#combinação = attitude["learning"].idxmax(),["rating","complaints"]
combinação
```

**3 - Qual é o nível médio de aprendizado para funcionários que receberam aumentos (raises) acima da média?**

```python
media_aprendizado = attitude["raises"].mean()
media_aprendizado
```

```python
acimaMedia_aprendizado = attitude[attitude["raises"] > media_aprendizado]
acimaMedia_aprendizado
```

# Desafio 7: Titanic

Definição do problema: Dado que você está trabalhando com um
grupo de historiadores que estão analisando o naufrágio do Titanic,
informe aos historiadores as seguintes questões:

-----------------

Código do dataset: titanic

-----------------

Dicas:
- atributo do DataFrame: shape
- Método map()
- Método groupby()

```python
titanic = pydataset.data("titanic")
```

```python
titanic.head()
```

**1. Qual a taxa de sobrevivência geral no navio?**

```python
if titanic["survived"].dtype == "object":
  titanic["survived"] = titanic["survived"].map({"yes":1, "no":0})
```

```python
taxaSobrevivencia = titanic["survived"].mean()
taxaSobrevivencia
```

**2. Qual a taxa de sobrevivência por classe (class)?**



```python
taxa = titanic.groupby("class")["survived"].mean().reset_index()
taxa
```

**3. Qual o sexo com maior taxa de sobrevivência?**

```python
taxaSobrevivencia_sexo = titanic.groupby("sex")["survived"].mean().reset_index()
taxaSobrevivencia_sexo
```

**4. Baseado nos dados informados. Qual a quantidade de adultos e
crianças (sobreviventes) e não sobreviventes?**

```python
adultoSobrevivente = titanic[(titanic["age"]=="adults")& (titanic["survived"]==1)].shape[0]
```

```python
adultoNobrevivente = titanic[(titanic["age"]=="adults")& (titanic["survived"]==0)].shape[0]
```

```python
criancaSobrevivente = titanic[(titanic["age"]=="child")& (titanic["survived"]==1)].shape[0]
```

```python
criancaNobrevivente = titanic[(titanic["age"]=="child")& (titanic["survived"]==0)].shape[0]
```

```python
print(f"A quantidade de adultos e crianças que sobreviveram é: {adultoSobrevivente} e {criancaSobrevivente}")
print(f"A quantidade de adultos e crianças que não sobreviveram é: {adultoNobrevivente} e {criancaNobrevivente}")
```
